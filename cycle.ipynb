```python
import pandas as pd
import requests
import json

# Fetch data from the URLs
url1 = "https://cycling.data.tfl.gov.uk/CyclingInfrastructure/data/points/cycle_parking.json"
url2 = "https://cycling.data.tfl.gov.uk/CyclingInfrastructure/data/points/restricted_point.json"

response1 = requests.get(url1)
response2 = requests.get(url2)

# Convert the response to JSON
data1 = response1.json()
data2 = response2.json()
```

```python
# Convert the JSON data to pandas DataFrame
df1 = pd.json_normalize(data1['features'])
df2 = pd.json_normalize(data2['features'])

# Clean the data (for example, handle missing values)
df1 = df1.fillna('Unknown')
df2 = df2.fillna('Unknown')
```

```python
from scipy.spatial.distance import cdist

# Sample data (replace with your actual data)
df1 = pd.DataFrame({
    'geometry.coordinates': [[-0.1278, 51.5074], [-0.1239, 51.5113], [-0.1198, 51.5152]]
})

df2 = pd.DataFrame({
    'geometry.coordinates': [[-0.1300, 51.5050], [-0.1250, 51.5125], [-0.1200, 51.5170]]
})

# Calculate distances using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    r = 6371  # Earth radius in km
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2) ** 2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return r * c

# Unpack coordinates and calculate distances
df1['latitude'], df1['longitude'] = zip(*df1['geometry.coordinates'])
df2['latitude'], df2['longitude'] = zip(*df2['geometry.coordinates'])

distances = pd.DataFrame(cdist(df1[['latitude', 'longitude']], df2[['latitude', 'longitude']], haversine),
                         columns=df2.index, index=df1.index)

# Add the minimum distance and corresponding point from df2 to df1
df1['min_distance'] = distances.min(axis=1)
df1['closest_point'] = df2.loc[distances.idxmin(axis=1), 'geometry.coordinates'].values.tolist()

print(df1)
```
