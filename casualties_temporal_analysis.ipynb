```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import geopandas as gpd
```

```python
# URL of the JSON file
url = "https://cycling.data.tfl.gov.uk/CyclingInfrastructure/data/lines/advanced_stop_line.json"

# Fetch the content from the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()
    
    # Convert JSON data to pandas DataFrame
    df_adv_stop_line = pd.DataFrame(data)
    
    # Export DataFrame to CSV
    df_adv_stop_line.to_csv('advanced_stop_line_data.csv', index=False)
    
    # Export DataFrame to Excel
    # df.to_excel('signal_data.xlsx', index=False)
    
    print("Data exported successfully.")
else:
    print("Failed to fetch data. Status code:", response.status_code)
```

```python
# Create a DataFrame
total_casualties = pd.DataFrame({
    "Year": ["2016", "2017", "2018", "2019", "2020", "2021", "2022"],
    "Total Casualties": [30270, 32567, 30591, 30007, 24345, 26671, 27207],
    "Total Casualties - Pedal Cycle": [4424, 4521, 4755, 4634, 4789, 5277, 5091],
    "Total Km Road Travel": [20083026279, 20055359109, 19992861706, 20264369690, 16613638242, 18453291284, 19089347093],
    "Total Km Pedal Cycle Travel": [506307274.6, 500504499.9, 518009547.1, 467836359.5, 697725718.5, 647245366.3, 605772496.9],
    "Total Casualties - Speed Limit <= 20 MPH": [2784, 6274, 7156, 8505, 8747, 11471, 12585],
    "Total Casualties - Speed Limit 30 MPH": [25013, 23138, 20374, 18581, 13340, 12831, 12268],
    "Total Casualties - Speed Limit 40 MPH": [1387, 1776, 1690, 1650, 1335, 1307, 1343],
    "Total Casualties - Speed Limit 50 MPH": [822, 975, 979, 910, 689, 795, 786],
    "Total Casualties - Speed Limit 60 MPH": [32, 88, 69, 65, 49, 48, 44],
    "Total Casualties - Speed Limit 70 MPH": [232, 315, 323, 292, 180, 205, 179]
})

# Calculate Total Road Accidents per Million Km Road Travel
total_casualties["Total Road Accidents per Million Km Road Travel"] = (total_casualties["Total Casualties"] / total_casualties["Total Km Road Travel"]) * 1000000
total_casualties["Total Casualties per Million Km Road Travel - Speed Limit <= 20 MPH"] = (total_casualties["Total Casualties - Speed Limit <= 20 MPH"] / total_casualties["Total Km Road Travel"]) * 1000000
total_casualties["Total Casualties - Pedal Cycle per Million Km Road Travel"] = (total_casualties["Total Casualties - Pedal Cycle"] / total_casualties["Total Km Pedal Cycle Travel"]) * 1000000

# Plot the DataFrame
fig, ax = plt.subplots()
ax.plot(total_casualties["Year"], total_casualties["Total Road Accidents per Million Km Road Travel"], color="g", marker="x")
ax.set_xlabel("Year")
ax.set_ylabel("Casualties per Million Km Road Travel", color="green")
ax2 = ax.twinx()
ax2.plot(total_casualties["Year"], total_casualties["Total Casualties per Million Km Road Travel - Speed Limit <= 20 MPH"], color="r", marker=".")
ax2.set_ylabel("Speed Limit <=20MPH Roads' Casualties per Million Km Road Travel", color="red")

# Add a title and labels
#plt.title("Count of Total Road Accidents per Million Km Road Travel by Year")
#plt.xlabel("Year")
#plt.ylabel("Total Road Accidents per Million Km Road Travel")

#fig.set_size_inches([])
fig.savefig("casualties and speed limit casualties normalised.png")
plt.show()
```
